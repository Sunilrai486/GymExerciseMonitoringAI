```python
import pandas as pd

# Assuming your DataFrame is named df_clusters
# Adjust the DataFrame variable accordingly
df_clusters = ...

def get_cluster_indices(df_clusters):
    cluster_columns = [col for col in df_clusters.columns if '_cluster' in col]

    # Create an empty dictionary to store cluster indices
    cluster_indices = {}

    # Iterate through each cluster column
    for cluster_col in cluster_columns:
        # Group the DataFrame by the cluster column
        grouped = df_clusters.groupby(cluster_col)

        # Retrieve indices for each cluster
        indices = {cluster: group.index.tolist() for cluster, group in grouped}
        
        # Add to the dictionary
        cluster_indices[cluster_col] = indices

    return cluster_indices

# Call the function to get the cluster indices
indices_dict = get_cluster_indices(df)

# Print or use the indices as needed
print(indices_dict)
```

```python
def get_indices_for_cluster(df_clusters, cluster_column, cluster_label):
        # Check if the specified cluster column exists in the DataFrame
        if cluster_column not in df_clusters.columns:
            print(f"Error: {cluster_column} not found in DataFrame.")
            return None

        # Filter rows based on the specified cluster label
        cluster_indices = df_clusters[df_clusters[cluster_column] == cluster_label].index.tolist()

        return cluster_indices
    
# Call the method to get indices for a specific cluster
cluster_column_name = 'left_knee_cluster'  # Replace with your actual cluster column name
cluster_label_value = 1  # Replace with your actual cluster label value
indices_for_cluster = get_indices_for_cluster(df, cluster_column_name, cluster_label_value)
indices_for_cluster
```

```python
def separate_non_continuous_numbers(numbers):
    clusters = []
    current_cluster = []

    for num in sorted(numbers):
        if not current_cluster or num == current_cluster[-1] + 1:
            current_cluster.append(num)
        else:
            clusters.append(current_cluster)
            current_cluster = [num]

    if current_cluster:
        clusters.append(current_cluster)

    return clusters

# Example usage:
input_numbers = indices_for_cluster
output_clusters = separate_non_continuous_numbers(input_numbers)
print(output_clusters)
```
